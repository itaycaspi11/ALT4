```python
import matplotlib.pyplot as plt
import pandas as pd
file_name = r"C:\Users\Student 1\Desktop\Microbittemprature.csv"
data = pd.read_csv(file_name)

#Updated file path - its working fine now.
```

```python
mean_temperature = round(data["Temperature"].mean(), 2)
min_temperature = data["Temperature"].min()
max_temperature = data["Temperature"].max()
median_temperature = data["Temperature"].median()
```

```python
print(f"\nStatistics:")
print(f"Mean: {mean_temperature}")
print(f"Min: {min_temperature}")
print(f"Max: {max_temperature}")
print(f"Median: {median_temperature}")
```

```python
plt.plot(data["Temperature"])
plt.axhline(mean_temperature, color = "red", linestyle = "--", label = f"Mean: {mean_temperature}" ) 
plt.title("Microbit Temperature Data")
plt.xlabel("Sample Number")
plt.legend()
plt.grid(True)
plt.show()

#I added quotation marks around "red" 
```
